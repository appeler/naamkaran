```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab/Naamkaran/data/fl_reg_data.csv")
```

```python
df = df[["name_first", "gender"]]
```

```python
df.head()
```

```python
df["gender"].value_counts()
```

```python
import numpy as np

# drop gender unknown
df["gender"].replace("U", np.nan, inplace=True)
df.dropna(subset=["gender"], inplace=True)
```

```python
df["gender"].value_counts()
```

```python
# Remove NA first/last
df.dropna(subset=["name_first"], inplace=True)
df = df.drop(df[df["name_first"].str.len() < 2].index)
df["name_first"] = df.name_first.str.strip().str.title()

df["name_first"] = df["name_first"].str.replace("[^a-zA-Z' -]", "", regex=True)

df.head()
```

```python
df.drop_duplicates(subset=["name_first"], inplace=True)
```

```python
df.shape
```

```python
df.to_csv("/content/drive/MyDrive/Colab/Naamkaran/data/fl_first_names.csv", index=False)
```
